```python
factory = StopWordRemoverFactory()
stopwords_id = factory.get_stop_words()
stopword = factory.create_stop_word_remover()
exclude = set(string.punctuation)
#lemma = Lemmatizer()
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
nlp = Indonesian()
nlp = spacy.blank('id')
```

```python
vectorizer = TfidfVectorizer(stop_words=stopwords_id)
```

```python
def remove_character(s):
    return "".join(i for i in s if ord(i) < 128)
```

```python
def clean(doc):
    clean_punc = ''.join(ch for ch in doc.lower() if ch not in exclude)
    clean_stopwords = stopword.remove(clean_punc)
    #stemm = stemmer.stem(clean_stopwords)
    teks = nlp(clean_stopwords)
    normalized = " ".join(token.lemma_ for token in teks)
    return normalized.strip()
```

```python
fields = ['judul','label']
dataset = pd.read_csv('djudul.csv',skipinitialspace=True,usecols=fields)
```

```python
train_clean_sentences = []
data = ''
for title in dataset.judul:
    clean_document = clean(title)
    data = data + clean_document
    print(clean_document)
    train_clean_sentences.append(clean_document)
x =vectorizer.fit_transform(train_clean_sentences)
y = dataset.label
for (i,item) in enumerate(y):
    if item == "SI":
        y[i] = 3
    elif item == "Jaringan":
        y[i] = 2
    elif item == "AI":
        y[i] = 1
    else:
        y[i] = 0
```

```python
k_range = range(1, 35)
k_score = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, x, y.astype('int'),cv=kf,scoring="accuracy")
    k_score.append(score.mean()*100)

print(k_score)

plt.plot(k_range, k_score)
plt.xlabel("K")
plt.ylabel("score K")
plt.show()
```

```python
abstract = []
for i in os.listdir("artikel1/"):
    if i.endswith(".txt"):
        file_abstract = open(os.path.join('artikel1/', i), 'r')
        abstract.append(remove_character(file_abstract.read()))

cv = CountVectorizer(ngram_range=(1, 1), stop_words=stopwords)
X = cv.fit_transform(abstract)
Xc = (X.T * X)
Xc.setdiag(0)

names = cv.get_feature_names()
df = pd.DataFrame(data=Xc.toarray(), columns=names, index=names)
df.to_csv("csv_data_text.csv", sep=",")
```
